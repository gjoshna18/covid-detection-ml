```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
```

```python
dataset=pd.read_csv('E:\project\AdaptiveHistogram.csv')
```

```python
X=dataset.iloc[:,0:625].values
y=dataset.iloc[:,625].values
```

```python
X
```

```python
y
```

```python
#Splitting training and testing data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#LDA for dimensionality reduction
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda=LDA(n_components=1)
```

```python
X_train_lda=lda.fit_transform(X_train,y_train)
X_test_lda=lda.transform(X_test)
```

```python
lda.n_components
```

```python
X_train_lda
```

```python
X_train_lda.shape
```

```python
X_test_lda
```

```python
X_test_lda.shape
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_lda,y_train)
```

```python
y_pred_logreg=logreg.predict(X_test_lda)
```

```python
y_pred_logreg
```

```python
conf_matrix = confusion_matrix(y_test,y_pred_logreg)
conf_matrix
```

```python
accuracy_score(y_test, y_pred_logreg)
```

```python
precision_score(y_test, y_pred_logreg,average='weighted')
```

```python
recall_score(y_test, y_pred_logreg,average='weighted')
```

```python
f1_score(y_test, y_pred_logreg, average='weighted')
```

```python
recall_score(y_test, y_pred_logreg, pos_label='Non-Covid')
```

```python
specificity=conf_matrix[1,1]/(conf_matrix[1,1]+conf_matrix[1,0])
```
