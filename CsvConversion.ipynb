```python
import numpy as np
from PIL import Image
import cv2
import os
```

```python
def convert(folder):
    IMG_DIR = 'E:/project/'+folder
    for img in os.listdir(IMG_DIR):
        img_array = cv2.imread(os.path.join(IMG_DIR,img), cv2.IMREAD_GRAYSCALE)
        img_pil = Image.fromarray(img_array)
        img_25x25 = np.array(img_pil.resize((25,25), Image.Resampling.LANCZOS))
        img_array = (img_25x25.flatten())
        img_array  = img_array.reshape(-1,1).T
        #print(img_array)
        with open('{}.csv'.format(folder), 'ab') as f:
            np.savetxt(f, img_array, delimiter=",")
```

```python
convert('NormalAdaptiveHistogram')
```

```python
convert('CovidAdaptiveHistogram')
```

```python
convert('NormalHistogramEqualized')
```
