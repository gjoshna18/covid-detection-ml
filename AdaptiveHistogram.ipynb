```python
#Processing the covid and normal iamges into Adaptive Histogram images and saving in folders
```

```python
import cv2
import os
from matplotlib import pyplot as plt
```

```python
def get_names(path):
    names = []
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            _, ext = os.path.splitext(filename)
            if ext in ['.png','.jpg']:
                names.append(filename)
    return names
```

```python
path='E:/project/'
```

```python
normalImages = get_names(path+'NormalImages')
```

```python
print(len(normalImages))
```

```python
covidImages = get_names(path+'CovidImages')
```

```python
print(len(covidImages))
```

```python
def saveImages(images,path1,path2):
    for image in images:
        img = cv2.imread('E:/project/'+path1+'{}'.format(image),0)
        #plt.imshow(img, 'gray')
        clahe = cv2.createCLAHE(clipLimit = 2, tileGridSize=(8,8))
        equ = clahe.apply(img)
        #plt.imshow(equ, 'gray')
        cv2.imwrite('E:/project/'+path2+'{}'.format(image),equ)
        #cv2.imwrite(f'E:/project/{path2}/{image}', equ)
        #break
```

```python
saveImages(normalImages,'NormalImages/','NormalAdaptiveHistogram/')
```

```python
saveImages(covidImages,'CovidImages/','CovidAdaptiveHistogram/')
```
