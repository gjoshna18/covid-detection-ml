```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('E:\project\AdaptiveHistogram.csv')
```

```python
X=dataset.iloc[:,0:625].values
y=dataset.iloc[:,625].values
```

```python
#Feature Scaling
#Standardize the data
#X_std=(x-mean(x))/std_dev(x)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X=StandardScaler()
sc_X=sc_X.fit(X_train)
```

```python
X_train=sc_X.transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
import pickle
```

```python
with open("sc_ahnor.pk","wb") as f:
    pickle.dump(sc_X, f)
```

```python
with open('sc_ahnor.pk', 'rb') as pickle_file:
    sc_X = pickle.load(pickle_file)
```

```python
sc_X
```

```python
X_train=sc_X.transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
#PCA for dimensionality reduction
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.90)
```

```python
pca_model=pca.fit(X_train)
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
import pickle
```

```python
with open("pca_ahnor.pk","wb") as f:
    pickle.dump(pca_model, f)
```

```python
with open('pca_ahnor.pk', 'rb') as pickle_file:
    pca_model = pickle.load(pickle_file)
```

```python
pca_model.n_components_
```

```python
X_train_pca=pca_model.transform(X_train)
X_test_pca=pca_model.transform(X_test)
```

```python
pca.n_components_
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_pca,y_train)
```

```python
y_pred_logreg=logreg.predict(X_test_pca)
```

```python
y_pred_logreg
```

```python
conf_matrix = confusion_matrix(y_test,y_pred_logreg)
conf_matrix
```

```python
accuracy_score(y_test, y_pred_logreg)
```

```python
precision_score(y_test, y_pred_logreg,average='weighted')
```

```python
recall_score(y_test, y_pred_logreg, average='weighted')
```

```python
f1_score(y_test, y_pred_logreg, average='weighted')
```

```python
recall_score(y_test, y_pred_logreg, pos_label='Non-Covid')
```
